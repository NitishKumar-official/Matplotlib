# bins, cumulative , histtype

```python
no = [43, 31, 17, 40, 34, 41, 17, 54, 30, 55, 57, 30, 42, 10, 36, 30, 35, 30, 44, 39, 55, 18, 37 ,12,
 57, 14, 51, 49, 36, 37, 24, 51, 56, 59, 47, 24, 13, 12, 27, 20, 15, 59, 45, 33, 37, 16, 45, 58,
 17, 13]
l=[10,20,30,40,50,60]
plt.hist(no,color="b",bins=l,edgecolor="r",cumulative=-1, histtype="step")

plt.title("Nitish RANA", fontsize=15)
plt.xlabel("Python", fontsize=15)
plt.ylabel("No..", fontsize=15)
plt.show()
```
