# Scatter Plot

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
No=[2,3,1,4,5,3,6]
plt.scatter(day,No)
plt.title("Scatterplot", fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No", fontsize=15)
plt.show()
```

### color , Size, blur(aplha->0-1)

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
No=[2,3,1,4,5,3,6]
colors = ["r","y","g","b","r","g","r",]
sizes = [500,200,100,50,48,63,700]
plt.scatter(day,No, c=colors, s=sizes, alpha=0.4)
plt.title("Scatterplot", fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No", fontsize=15)
plt.show()
```

### change pointer, edgecolor, linewidth

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
No=[2,3,1,4,5,3,6]
colors = ["r","y","g","b","r","g","r",]
sizes = [500,200,100,50,48,63,700]
plt.scatter(day,No, c=colors, s=sizes,marker="*", edgecolors="g", linewidths=2)
plt.title("Scatterplot", fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No", fontsize=15)
plt.show()
```

# ColorMap, colorBar

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
No=[2,3,1,4,5,3,6]
colors = [10,49,30,29,56,20,3]
sizes = [500,200,100,50,48,63,700]
plt.scatter(day,No, c=colors, s=sizes,cmap="BrBG")
t=plt.colorbar()
t.set_label("ColorBar", fontsize=15)
plt.title("Scatterplot", fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No", fontsize=15)
plt.show()
```
