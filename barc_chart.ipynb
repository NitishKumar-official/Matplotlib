# Bar chart

```python
import matplotlib.pyplot as plt
x = ["python","c", "c++", "java"]
y = [85,70,60,82]
c=["y","b","r","g"]
plt.bar(x,y, width=0.4,color=c,align="edge",edgecolor="black",linewidth=3, linestyle=":",alpha=0.5, label="NitishRana")  #align=centre  ->by default
plt.legend() #->   label="NitishRana" -> legend ka use isko show krwane ke liye krte hain
plt.xlabel("Programming Language", fontsize=20)
plt.ylabel("popularity",fontsize=20)
plt.title("porgramming language popularity",fontsize=20)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = ["python","c", "c++", "java"]
y = [85,70,60,82]
z = [20,30,40,50]
plt.xlabel("Programming Language", fontsize=20)
plt.ylabel("popularity",fontsize=20)
plt.title("porgramming language popularity",fontsize=20)
plt.bar(x,y, width=0.4,color="r",label="NitishRana")
plt.bar(x,z, width=0.4,color="y",label="NitishRanaG")
plt.legend() 
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = ["python","c", "c++", "java"]
y = [85,70,60,82]
z = [20,30,40,50]
p = np.arange(len(x))
width=0.2
p1 = [j+width for j in p]

plt.xlabel("Programming Language", fontsize=20)
plt.ylabel("popularity",fontsize=20)
plt.title("porgramming language popularity",fontsize=20)

plt.bar(p,y, width,color="r",label="NitishRana")
plt.bar(p1,z, width,color="y",label="NitishRanaG")

plt.xticks(p+width/2, x,rotation=20)
plt.legend() 
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = ["python","c", "c++", "java"]
y = [85,70,60,82]
z = [20,30,40,50]
p = np.arange(len(x))
width=0.2
p1 = [j+width for j in p]

plt.xlabel("Programming Language", fontsize=20)
plt.ylabel("popularity",fontsize=20)
plt.title("porgramming language popularity",fontsize=20)

plt.barh(p,y, width,color="r",label="NitishRana")
plt.barh(p1,z, width,color="y",label="NitishRanaG")

plt.xticks(p+width/2, x,rotation=20)
plt.legend() 
plt.show()
```
